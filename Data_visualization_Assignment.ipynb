Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

Ans.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset with correlated features
size = 100  # Number of samples
feature_1 = np.random.normal(50, 10, size)
feature_2 = feature_1 * 0.8 + np.random.normal(0, 5, size)  # Correlated with feature_1
feature_3 = feature_1 * -0.5 + feature_2 * 0.3 + np.random.normal(0, 5, size)  # Correlated with both
feature_4 = np.random.normal(60, 15, size)  # Less correlated feature

# Create DataFrame
df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Compute correlation matrix
correlation_matrix = df.corr()

# Visualize correlation matrix using heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute z values based on the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Create DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create 3D surface plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Surface Plot of z = sin(sqrt(x^2 + y^2))")

# Show plot
plt.show()
```

Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

Ans.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# Enable output display in Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)
num_points = 100
x = np.random.uniform(0, 10, num_points)
y = np.random.uniform(0, 10, num_points)
sizes = np.random.randint(10, 50, num_points)  # Marker sizes
colors = np.random.choice(["red", "blue", "green", "purple", "orange"], num_points)  # Random colors

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Convert DataFrame to ColumnDataSource (for Bokeh)
source = ColumnDataSource(df)

# Create a Bokeh figure
p = figure(title="Bokeh Scatter Plot with Varying Sizes and Colors",
           x_axis_label="X values", y_axis_label="Y values", width=700, height=400)

# Add scatter plot
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6, legend_field="colors")

# Customize legend
p.legend.title = "Colors"
p.legend.location = "top_right"

# Show plot
show(p)
```

Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

Ans.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral4

# Enable output display in Jupyter Notebook
output_notebook()

# Define dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Convert to ColumnDataSource for Bokeh
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruit", y_axis_label="Count", width=700, height=400)

# Add bar chart with colors
p.vbar(x='fruits', top='counts', width=0.6, source=source,
       legend_field="fruits", color=factor_cmap('fruits', palette=Spectral4, factors=fruits))

# Customize the legend
p.legend.title = "Fruit Type"
p.legend.location = "top_right"

# Show plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh output in Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)  # 1000 random values from a normal distribution

# Compute histogram bins
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Data Distribution", x_axis_label="Value", y_axis_label="Frequency", width=700, height=400)

# Add quad glyphs to create a histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="black", alpha=0.7)

# Show plot
show(p)
```

Q5.  Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

Ans.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.transform import transform
import numpy as np
import pandas as pd

# Enable Bokeh output in Jupyter Notebook
output_notebook()

# Generate heatmap data
data_heatmap = np.random.rand(10, 10)

# Generate x and y grid
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten data for Bokeh
df = pd.DataFrame({'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()})

# Convert DataFrame to ColumnDataSource
source = ColumnDataSource(df)

# Define color mapper
mapper = LinearColorMapper(palette="Viridis256", low=df['value'].min(), high=df['value'].max())

# Create a Bokeh figure
p = figure(title="Bokeh Heatmap", x_axis_label="X Axis", y_axis_label="Y Axis", width=600, height=500)

# Add heatmap using rect glyph
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=transform('value', mapper), line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
